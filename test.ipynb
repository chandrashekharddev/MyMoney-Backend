```python
import sqlite3
```

```python
conn = sqlite3.connect('example.db')
cursor = conn.cursor()
```

```python
q1=""" CREATE TABLE IF NOT EXISTS expenses (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    date TEXT NOT NULL,
    category TEXT NOT NULL,
    amount REAL NOT NULL,
    description TEXT); """
```

```python
cursor.execute(q1)
conn.commit()
```

```python
q2 = """ INSERT INTO expenses (date, category, amount, description)
         VALUES ('2024-06-01', 'Groceries', 150.75, 'Weekly grocery shopping'); """
```

```python
cursor.execute(q2)
conn.commit()
```

```python
q3 = "SELECT * FROM expenses;"
cursor.execute(q3)
rows = cursor.fetchall()
for row in rows:
    print(row)
```
